```python
import numpy as np
```

```python
n = 5
A = np.random.random((n,n))
A_iter = A.copy()
```

```python
D1 = np.eye(n) # To scale the rows
D2 = np.eye(n) # To scale the columns
```

```python
D1.dot(A.dot(D2))
```

```python
n_iterations = 30
for i in range(n_iterations):
    column_sums = A_iter.sum(axis=0)
    row_sums = A_iter.sum(axis=1)
    if (i % 2 )== 0:
        D1 *= np.diag(1.0 / row_sums)
    else:
        D2 *= np.diag(1.0 / column_sums)
    A_iter = D1.dot(A.dot(D2)) 
    print(A_iter.sum(axis=0), A_iter.sum(axis=1))
```

```python
A_normalised = D1.dot(A.dot(D2))
print(A_normalised)
```

```python
print(A_normalised.sum(0), A_normalised.sum(1))
```
